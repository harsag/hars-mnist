```python
#hide
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
from ipywidgets import Image

from ipywidgets import ColorPicker, IntSlider, link, AppLayout, HBox

from ipycanvas import Canvas, RoughCanvas, hold_canvas
```

```python
width = 280
height = 280
```

```python
from math import pi
canvas = Canvas(width=width, height=height)

drawing = False
position = None

def on_mouse_down(x, y):
    global drawing
    global position

    drawing = True
    position = (x, y)


def on_mouse_move(x, y):
    global drawing
    global position

    if not drawing:
        return

    with hold_canvas(canvas):
        canvas.fill_arc(position[0], position[1], 6, 0, 2 * pi)
    #    canvas.stroke_line(position[0], position[1], x, y)

    position = (x, y)

def on_mouse_up(x, y):
    global drawing
    global position

    drawing = False

    #with hold_canvas(canvas):
    #    canvas.stroke_line(position[0], position[1], x, y)

def handle_touch_move(fingers_locations):
    # Draw circles where fingers are located
    for finger_location in fingers_locations:
        with hold_canvas(canvas):
            canvas.fill_arc(finger_location[0], finger_location[1], 6, 0, 2 * pi)

canvas.on_mouse_down(on_mouse_down)
canvas.on_mouse_move(on_mouse_move)
canvas.on_mouse_up(on_mouse_up)
canvas.on_touch_move(handle_touch_move)

canvas.stroke_style = "#749cb8"

canvas.sync_image_data = True
canvas.fill_style = 'black'
canvas.fill_rect(0, 0, 280, 280)
canvas.fill_style = 'white'
```

```python
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
#learn_inf = load_learner(path/'export.pkl', cpu=True)
```

```python
btn_run = widgets.Button(description='Identify')
btn_clr = widgets.Button(description='Clear')
```

```python
def on_click_identify(change):
    #img = PILImage.create(btn_upload.data[-1])
    img = PILImage.create(canvas.get_image_data())
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(28,28))
    #pred,pred_idx,probs = learn_inf.predict(img)
    #lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'
    lbl_pred.value = f'Prediction: It is a . . . umm . . . I dont know... '
```

```python
btn_run.on_click(on_click_identify)
```

```python
def on_click_clear(change):
    canvas.clear()
```

```python
btn_clr.on_click(on_click_clear)
```

```python
display(VBox([widgets.Label('Please draw any digit (0-9) here . . .'), widgets.Label('Kindly draw it very very slowly....'), widgets.Label('Because slow is beautiful :)'), canvas, btn_run, btn_clr, out_pl, lbl_pred]))
```
